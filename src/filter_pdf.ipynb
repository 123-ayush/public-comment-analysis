```python
import pandas as pd
import re

pdf_path = r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\pdf_extracted_comments.csv"
output_path = r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\pdf_filtered_200.csv"

df = pd.read_csv(pdf_path)

# Keep sentences ONLY between 10â€“35 words (remove long reports)
df["word_count"] = df["comment_text"].apply(lambda x: len(str(x).split()))
df = df[(df.word_count >= 10) & (df.word_count <= 35)]

# Remove duplicates & shuffle
df = df.drop_duplicates(subset=["comment_text"])
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Keep only top 200 clean entries
df = df.head(200)

df.to_csv(output_path, index=False)
print("Filtered PDF dataset saved:", output_path, "| Rows:", len(df))
```
