```python
import pandas as pd
import os

base_path = r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset"

files = [
    "policy_comment_dataset_500.csv",
    "synthetic_policy_comments_1200.csv",
    "playstore_policy_reviews.csv",
    "short_comments_dataset.csv",
    "pdf_filtered_200.csv"
]

dfs = [pd.read_csv(os.path.join(base_path, f)) for f in files]

final_df = pd.concat(dfs, ignore_index=True)

final_df.drop_duplicates(subset=["comment_text"], inplace=True)
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

output = r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\final_clean_dataset.csv"
final_df.to_csv(output, index=False)

print("ðŸŽ‰ Final cleaned dataset created successfully!")
print("Rows:", len(final_df))
```
