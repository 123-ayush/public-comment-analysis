```python
import pandas as pd
import os

support = [
    "good policy", "love this policy", "very helpful", "great initiative",
    "excellent step", "happy with this policy", "like the policy",
    "really support this", "this helps a lot", "amazing decision"
]

oppose = [
    "bad policy", "policy sucks", "useless plan", "waste of money",
    "terrible decision", "failed badly", "not helpful at all",
    "strongly oppose", "worst policy", "very disappointed"
]

neutral = [
    "not sure", "depends", "could be better", "need improvement",
    "average policy", "neutral view", "unclear impact", "mixed reaction"
]

rows = []
for txt in support:
    rows.append({"comment_text": txt, "stance_label": "Support"})
for txt in oppose:
    rows.append({"comment_text": txt, "stance_label": "Oppose"})
for txt in neutral:
    rows.append({"comment_text": txt, "stance_label": "Neutral"})

df = pd.DataFrame(rows)

os.makedirs("dataset", exist_ok=True)
df.to_csv(r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\short_comments_dataset.csv", index=False)

print("Created short dataset: dataset/short_comments_dataset.csv")
print("Rows:", len(df))
```
