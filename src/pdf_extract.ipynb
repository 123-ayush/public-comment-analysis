```python
import fitz  # PyMuPDF
import os
import re
import pandas as pd

print("üìÑ Extracting PDF stance comments...")

pdf_files = [
    "An Analysis of Problems of Sukanya Samriddhi Account Scheme.pdf",
    "Impact Analysis of the NPEGEL A Study of Udalguri District of the State of Assam.pdf",
    "Sample Survey and Study under Section 12(1) (C) of the RTE Act, 2009.pdf",
    "Evaluation of Kasturba Gandhi Balika Vidyalayas (KGBVs) Scheme.pdf",
    "State-Survey-of-School-Meal-Programs-in-India-Report-with-Annexes.pdf",
    "SSG-2022-report.pdf",
    "Dialnet-PublicOpinionOnTheNewEducationPolicy2020-8271400.pdf"
]

support_kw = ["improved", "effective", "successful", "beneficial", "positive", "progress"]
oppose_kw = ["problem", "poor", "failed", "corruption", "issue", "lack", "ineffective", "slow"]
neutral_kw = ["depends", "unclear", "mixed", "average", "neutral", "uncertain"]

rows = []

for file_name in pdf_files:
    pdf_path = os.path.join(r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\PDFs", file_name)
    if not os.path.exists(pdf_path):
        print("‚ùå Not found:", file_name)
        continue

    print("üîç Processing:", file_name)
    doc = fitz.open(pdf_path)

    text = ""
    for page in doc:
        text += page.get_text()

    # Split into sentences using punctuation
    sentences = re.split(r'(?<=[.!?]) +', text)

    for s in sentences:
        s_clean = " ".join(s.split())

        if len(s_clean) < 40 or len(s_clean) > 250:
            continue

        stance = "Neutral"
        lower = s_clean.lower()

        if any(k in lower for k in support_kw):
            stance = "Support"
        elif any(k in lower for k in oppose_kw):
            stance = "Oppose"
        elif any(k in lower for k in neutral_kw):
            stance = "Neutral"

        rows.append({
            "file_name": file_name,
            "comment_text": s_clean,
            "stance_label": stance
        })

df = pd.DataFrame(rows)

# Remove duplicates
df.drop_duplicates(subset=["comment_text"], inplace=True)

# Shuffle
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save output
os.makedirs("dataset", exist_ok=True)
output = r"C:\Users\Ayush Ahlawat\OneDrive\Documents\Public Comment Analysis\public-comment-analysis\dataset\pdf_extracted_comments.csv"
df.to_csv(output, index=False)

print("\nüéâ Extraction Completed Successfully!")
print("üíæ Saved to:", output)
print("üìù Rows Extracted:", len(df))
```
